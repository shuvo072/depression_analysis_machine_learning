```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
# reading data
data = pd.read_csv('Depression and Happiness Factor Analysis.csv')
# remove null
data = data.dropna(axis=0)
print(data.shape)
data.drop(['Timestamp'],axis=1,inplace=True)
data.head()
```

```python
obj_col = [col for col in data.columns if data[col].dtypes=='object']
obj_col
```

```python
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder
encoder = LabelEncoder()
for col_name in obj_col:
  data[col_name] = encoder.fit_transform(data[col_name])
data
```

```python
corr_matrix = data.corr()
corr_matrix
```

```python

corr_matrix['How are you feeling right now?'].sort_values(ascending = False)
```

```python
## define target and features
data.head()
```

```python
print(data.shape)
```

```python

y = data['How are you feeling right now?']
X = data.drop(['How are you feeling right now?',
                       'Gender', 'Age', 'Relationship status',
                       'Which year are you in?','Have you used any social media within the last 6 hours?'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=1)

```

```python
print(X_train.shape, X_valid.shape)
```

```python
model = DecisionTreeClassifier(max_depth=4)
model = model.fit(X_train, y_train)
```

```python

model.score(X_train,y_train)
```

```python
model.score(X_valid,y_valid)
```
